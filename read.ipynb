```python
import pandas as pd
import numpy as np
import random
import string
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(50, 100)
```

```python
df=pd.DataFrame(np.random.randint(50, 100, size=(10,4)), index=[id() for i in range(10)], columns=['국어','영어','수학','사회'])
```

```python
df.iloc[[0],:]   #iloc는 index로 가져올 때
```

```python
# 과학이 추가되었습니다. 기존 학생들에게 각 랜덤으로 추가하시오
df.loc[:,'과학'] = pd.Series([score() for i in range(10)], index=df.index)
df
```

```python
# 각 학생들 개인별 총점을 표현하는 컬럼을 추가하세요.
df['총점'] = df.sum(axis=1)
df
# df.drop(columns=['','개인별총점','개인총점'])    #열 삭제
# df.drop(index=[10,11,12,13])     #행 삭제
```
